```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python

# Definição de parâmetros
n = 300

# Série de tendência// processo deterministico
X = 1000 + 0.1 * np.arange(1, n+1)

# Plotagem da série de tendência
plt.figure(figsize=(10, 4))
plt.plot(X)
plt.title('Série de Tendência')
plt.show()
```

```python

# Série de sazonalidade (senoide)//serie ciclica
S = 20 * np.sin((2 * np.pi / 12) * np.arange(1, n+1))

# Plotagem da série de sazonalidade
plt.figure(figsize=(10, 4))
plt.plot(S)
plt.title('Série de Sazonalidade')
plt.show()
```

```python

# Série com tendência e sazonalidade
Y = S + X

# Plotagem da série com tendência e sazonalidade
plt.figure(figsize=(10, 4))
plt.plot(Y)
plt.title('Série com Tendência e Sazonalidade')
plt.show()
```

```python
# Ruído branco
# 0 é a media e 5 a variancia
E = np.random.normal(0, 5, n)

# Plotagem do ruído branco
plt.figure(figsize=(10, 4))
plt.plot(E)
plt.title('Ruído Branco')
plt.show()
```

```python

# Série completa (tendência + sazonalidade + ruído)
Z = S + X + E

# Criando um objeto DataFrame para manipulação mais fácil
date_range = pd.date_range(start='1990-01-01', periods=n, freq='M')
Z_df = pd.DataFrame({'Z': Z}, index=date_range)

# Plotagem da série completa
plt.figure(figsize=(10, 4))
plt.plot(Z_df)
plt.title('Série Completa')
plt.show()
```

```python

# Decomposição da série em tendência, sazonalidade e componente residual
decomposition = seasonal_decompose(Z_df, model='additive')

# Plotagem da decomposição
decomposition.plot()
plt.show()
```
