```python
with open('/content/docker_stats.txt', 'r', encoding='UTF-16') as file:
    lines = file.readlines()
    print(lines)
```

```python
# tratamento do txt

cleaned_lines = [line.replace('\x00', '').strip() for line in lines if line]

cleaned_lines = cleaned_lines[1:]

pattern = r'\s{2,}'
split_lines = [re.split(pattern, line) for line in cleaned_lines]

# Ajustar o cabealho do DataFrame
header = ['CONTAINER_ID', 'NAME', 'CPU', 'MEM_USAGE_LIMIT',  'MEM_PERCENTAGE', 'NET_IO', 'BLOCK_IO', 'PIDS']

# Criar DataFrame
df = pd.DataFrame(split_lines, columns=header)

df = df[~df['CONTAINER_ID'].str.contains('CONTAINER ID')]
df = df[~df['CPU'].str.contains('CPU %')]



```

```python
import numpy as np

df['CPU'] = df['CPU'].replace('--', np.nan)

df['CPU'] = df['CPU'].str.rstrip('%').astype(float)

# Calcular as cores
cores = plt.cm.viridis(df['CPU'] / max(df['CPU']))

plt.figure(figsize=(12, 8))
bars = plt.barh(df['NAME'], df['CPU'], color=cores)

for bar in bars:
    width = bar.get_width()
    if width > 5:
        plt.text(width, bar.get_y() + bar.get_height()/2, f'{width:.1f}%', va='center')

plt.xlabel('Utilizao de CPU (%)', fontsize=12)
plt.ylabel('Nome do Continer', fontsize=12)
plt.title('Utilizao de CPU por Continer', fontsize=16)
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```

```python
import matplotlib.pyplot as plt

df['CPU'] = df['CPU'].astype(str).str.rstrip('%').astype(float)

plt.figure(figsize=(15, 6))
for name, group in df.groupby('NAME'):
    plt.plot(group.index, group['CPU'], label=name)

plt.xlabel('Quant. Dados', fontsize=12)
plt.ylabel('Utilizao de CPU (%)', fontsize=12)
plt.title('Utilizao de CPU por Continer', fontsize=16)
plt.legend(loc='upper left', bbox_to_anchor=(0.78, 1), fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)


plt.savefig('utilizacao_cpu_containers.pdf', format='pdf', bbox_inches='tight')

plt.show()
```

```python
import matplotlib.pyplot as plt

df['CPU'] = df['CPU'].astype(str).str.rstrip('%').astype(float)


df['MEM_PERCENTAGE'] = df['MEM_PERCENTAGE'].replace('--', np.nan)
df['MEM_PERCENTAGE'] = df['MEM_PERCENTAGE'].astype(str).str.rstrip('%').astype(float)

avg_cpu_per_container = df.groupby('NAME')['CPU'].mean()

avg_mem_per_container = df.groupby('NAME')['MEM_PERCENTAGE'].mean()

total_cpu_over_time = df['CPU'].sum()

total_mem_over_time = df['MEM_PERCENTAGE'].sum()

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
avg_cpu_per_container.plot(kind='bar', rot=45, color='skyblue')
plt.title('Utilizao Mdia da CPU por Continer')
plt.xlabel('Continer')
plt.ylabel('Utilizao Mdia da CPU (%)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.subplot(2, 2, 2)
avg_mem_per_container.plot(kind='bar', rot=45, color='lightgreen')
plt.title('Utilizao Mdia da Memria por Continer')
plt.xlabel('Continer')
plt.ylabel('Utilizao Mdia da Memria (%)')
plt.grid(axis='y', linestyle='--', alpha=0.5)


plt.tight_layout()


#plt.savefig('utilizacao_media.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['NET_IO'] = df['NET_IO'].replace('--', np.nan)
df['NET_IO'] = df['NET_IO'].str.replace(r'\D', '', regex=True).astype(float) / 1024

avg_net_io_per_container = df.groupby('NAME')['NET_IO'].mean()

total_net_io_over_time = df['NET_IO'].sum()

plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
avg_net_io_per_container.plot(kind='bar', rot=45, color='skyblue')
plt.title('Utilizao Mdia de Rede por Continer')
plt.xlabel('Continer')
plt.ylabel('Utilizao Mdia de Rede (kB)')
plt.grid(axis='y', linestyle='--', alpha=0.5)


plt.tight_layout()


plt.savefig('utilizacao_media_rede_por_container.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

total_cpu_per_container = df.groupby('NAME')['CPU'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 8))
explode = [0.1 if i == 0 else 0 for i in range(len(total_cpu_per_container))]
plt.pie(total_cpu_per_container, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Distribuio da Utilizao de CPU por Continer')
plt.axis('equal')
plt.show()


total_mem_per_container = df.groupby('NAME')['MEM_PERCENTAGE'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 8))
explode = [0.1 if i == 0 else 0 for i in range(len(total_mem_per_container))]
plt.pie(total_mem_per_container, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title('Distribuio da Utilizao de Memria por Continer')
plt.axis('equal')
plt.legend(total_mem_per_container.index, loc="best")


plt.savefig('distribuio_da_utilizacao.pdf', format='pdf', bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

total_cpu_per_container = df.groupby('NAME')['CPU'].sum().sort_values(ascending=False)

total_mem_per_container = df.groupby('NAME')['MEM_PERCENTAGE'].sum().sort_values(ascending=False)

fig, axs = plt.subplots(1, 2, figsize=(15, 8))

explode_cpu = [0.1 if i == 0 else 0 for i in range(len(total_cpu_per_container))]
axs[0].pie(total_cpu_per_container, autopct='%1.1f%%', startangle=140, explode=explode_cpu)
axs[0].set_title('Distribuio da Utilizao de CPU por Continer')
axs[0].axis('equal')

explode_mem = [0.1 if i == 0 else 0 for i in range(len(total_mem_per_container))]
axs[1].pie(total_mem_per_container, autopct='%1.1f%%', startangle=140, explode=explode_mem)
axs[1].set_title('Distribuio da Utilizao de Memria por Continer')
axs[1].axis('equal')

plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.3)

axs[1].legend(total_mem_per_container.index, loc="best")

plt.savefig('distribuio_da_utilizacao.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
